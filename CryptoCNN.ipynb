```python
df_arr = []
for i in range(len(csv_array)):
    df_arr.append(pd.read_csv(PATH + csv_array[i]))
print(len(df_arr))
```

```python
ohlc_arr = []
for i in range(len(df_arr)):
    df_arr[i]['date'] = pd.to_datetime(df_arr[i]['date'])
    df_arr[i]['date'] = mdates.date2num(df_arr[i]['date'])
    ohlc_arr.append(df_arr[i][['date','open','high','low','close']])
print(len(ohlc_arr))
```

```python
f1, ax = plt.subplots(figsize = (10,5))

# plot Ethereum
candlestick_ohlc(ax, ohlc_ETH.values, width=.6, colorup='green', colordown='red')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.savefig('eth.png')
```
