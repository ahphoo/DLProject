```python
%matplotlib inline
import numpy as np
import pandas as pd
import pandas_datareader as datareader
import datetime
import matplotlib.pyplot as plt
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates
import os
```

```python
PATH = "C:\\Users\\PLSDONTCORRUPT\\Desktop\\CryptoData\\"
```

```python
csv_array = os.listdir(PATH)
for i in range(len(csv_array)):
    if(csv_array[i] == 'NEO_USD.csv'):
        print('Neo ' + str(i))
    if(csv_array[i] == 'XMR_USD.csv'):
        print('XMR ' + str(i))
    if(csv_array[i] == 'ETH_USD.csv'):
        print('ETH ' + str(i))
    if(csv_array[i] == 'LTC_USD.csv'):
        print('LTC ' + str(i))
print(csv_array)
```

```python
df_arr = []
for i in range(len(csv_array)):
    df_arr.append(pd.read_csv(PATH + csv_array[i]))
print(len(df_arr))
```

```python
ohlc_arr = []
for i in range(len(df_arr)):
    df_arr[i]['date'] = pd.to_datetime(df_arr[i]['date'])
    df_arr[i]['date'] = mdates.date2num(df_arr[i]['date'])
    ohlc_arr.append(df_arr[i][['date','open','high','low','close']])
print(len(ohlc_arr))
```

```python
f1, ax = plt.subplots(figsize = (10,5))

# plot Ethereum
candlestick_ohlc(ax, ohlc_arr[32].values, width=.6, colorup='green', colordown='red')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.show()
```
