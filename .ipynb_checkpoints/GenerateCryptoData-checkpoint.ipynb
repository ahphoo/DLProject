```python
%matplotlib inline
import numpy as np
import pandas as pd
import pandas_datareader as datareader
import datetime
import matplotlib.pyplot as plt
from mpl_finance import candlestick_ohlc
import matplotlib.dates as mdates
import os
```

```python
PATH = "C:\\Users\\PLSDONTCORRUPT\\Desktop\\CryptoData\\"
```

```python
csv_array = os.listdir(PATH)
for i in range(len(csv_array)):
    if(csv_array[i] == 'NEO_USD.csv'):
        print('Neo ' + str(i))
    if(csv_array[i] == 'XMR_USD.csv'):
        print('XMR ' + str(i))
    if(csv_array[i] == 'ETH_USD.csv'):
        print('ETH ' + str(i))
    if(csv_array[i] == 'LTC_USD.csv'):
        print('LTC ' + str(i))
print(csv_array)
```

```python
df_ETH = pd.read_csv(PATH + csv_array[32])
df_ETH.head(143214)
```

```python
# Generate 12 plots of Ethereum (3 months each from 2015-2018)
df_ETH_arr = []

# First year only has 1 plot
df_1 = df_ETH[(df_ETH['date'] >= '2015-08-00') & (df_ETH['date'] <= '2015-12-31')]
df_ETH_arr.append(df_1)

for i in range(2):
    year = 2016 + i
    df_1 = df_ETH[(df_ETH['date'] >= (str(year) + '-01-00')) & (df_ETH['date'] <= (str(year) + '-04-00'))]
    df_2 = df_ETH[(df_ETH['date'] >= (str(year) + '-04-00')) & (df_ETH['date'] <= (str(year) + '-07-00'))]
    df_3 = df_ETH[(df_ETH['date'] >= (str(year) + '-07-00')) & (df_ETH['date'] <= (str(year) + '-10-00'))]
    df_4 = df_ETH[(df_ETH['date'] >= (str(year) + '-10-00')) & (df_ETH['date'] <= (str(year) + '-12-31'))]
    df_ETH_arr.append(df_1)
    df_ETH_arr.append(df_2)
    df_ETH_arr.append(df_3)
    df_ETH_arr.append(df_4)

# Last year only has 3 plots
df_1 = df_ETH[(df_ETH['date'] >= '2018-01-00') & (df_ETH['date'] <= '2018-04-00')]
df_2 = df_ETH[(df_ETH['date'] >= '2018-04-00') & (df_ETH['date'] <= '2018-07-00')]
df_3 = df_ETH[(df_ETH['date'] >= '2018-07-00') & (df_ETH['date'] <= '2018-09-30')]
df_ETH_arr.append(df_1)
df_ETH_arr.append(df_2)
df_ETH_arr.append(df_3)
```

```python
print(len(df_ETH_arr))
```

```python
print(df_ETH_arr[0])
```

```python
ohlc_arr = []
for i in range(len(df_ETH_arr)):
    df_ETH_arr[i]['date'] = pd.to_datetime(df_ETH_arr[i]['date'])
    df_ETH_arr[i]['date'] = df_ETH_arr[i].date.map(mdates.date2num)
    ohlc_arr.append(df_ETH_arr[i][['date','open','high','low','close']])
print(len(ohlc_arr))
```

```python
f1, ax = plt.subplots(figsize = (10,5))

# plot Ethereum
candlestick_ohlc(ax, ohlc_arr[0].values, width=.6, colorup='green', colordown='red')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.show()
```
