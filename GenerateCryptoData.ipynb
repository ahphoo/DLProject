```python
ohlc_arr = []
for i in range(len(df_ETH_arr)):
    df_ETH_arr[i]['date'] = pd.to_datetime(df_ETH_arr[i]['date'])
    df_ETH_arr[i]['date'] = df_ETH_arr[i].date.map(mdates.date2num)
    ohlc_arr.append(df_ETH_arr[i][['date','open','high','low','close']])
print(len(ohlc_arr))
```

```python
f1, ax = plt.subplots(figsize = (10,5))

# plot Ethereum
candlestick_ohlc(ax, ohlc_arr[0].values, width=.6, colorup='black', colordown='black')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.show()
```

```python
for i in range(len(ohlc_arr)):
    f1, ax = plt.subplots(figsize = (10,5))

    # export Ethereum plot to PNG file
    candlestick_ohlc(ax, ohlc_arr[i].values, width=.6, colorup='black', colordown='black')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    # plt.show()
    plt.savefig("ohlc_arr[" + str(i) + "]" + ".png")
```
